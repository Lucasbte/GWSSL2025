```python
#create a table with the event's real positions, temperature and add speed of sound

def calc_speed_of_sound(temperature=20):
    """
    Calculate speed of sound in air, in meters per second

    Calculate speed of sound for a given temperature
    in Celsius (Humidity has a negligible
    effect on speed of sound and so this functionality
    is not implemented)

    Args:
        temperature: ambient air temperature in Celsius

    Returns:
        the speed of sound in air in meters per second
    """
    return 331.3 * np.sqrt(1 + float(temperature) / 273.15)

carol = {
    "#": list(range(1, 21)),
    "X": [10.9091, 14.5455, 18.1818, 21.8182, 25.4545, 29.0909, 32.7273, 36.3636, 40, 43.6364,
          47.2727, 50.9091, 54.5455, 58.1818, 61.8182, 65.4545, 69.0909, 72.7273, 76.3636, 80],
    "Y": [73.7426, 4.34896, 0.337676, 37.7563, 73.9322, 79.2031, 54.2929, 23.0378, 9.53521, 19.1982,
          37.5874, 45.2589, 34.5017, 14.1586, 0.0993164, 0.966436, 11.6436, 19.1036, 14.403, 0],
    "T": [23.3, 24, 23.5, 24, 23.3, 22.7, 22.7, 23, 23, 23.3,
          23, 22.4, 22.6, 22.8, 23, 23, 22.7, 22.8, 22.8, 22.6]
}

postemp = pd.DataFrame(carol)
postemp = postemp.set_index("#")
postemp["SoS"] = [calc_speed_of_sound(e) for e in carol["T"]]

postemp
```

```python
#initialize a SynchronizedRecorderArray object with the ARU coordinates

from opensoundscape.localization import SynchronizedRecorderArray
```

```python
#read a table of detections for the files

detections = pd.read_csv("/content/Recordings/detections.csv")

import pytz
from datetime import datetime, timedelta

#add information about the real-world timestamps, if the audio files don't have
#start timestamps embedded in the metadata
#in this case, when we synchronized the audio files we trimmed all of them to start
#at the same start time, and we can add that here

local_timestamp = datetime(2023, 3, 16, 4, 1, 16)
local_timezone = pytz.timezone("Pacific/Auckland")
detections["start_timestamp"] = [
    local_timezone.localize(local_timestamp) + timedelta(seconds=s)
    for s in detections["start_time"]
]

# set four columns as a multi-index, to match format expected for localization

detections = detections.set_index(["file", "start_time", "end_time", "start_timestamp"])
display(detections)
```

##Localization

```python
#loop for obtaining best obtained position for every event

results = {}

for i in range(20):
  detection = detections.iloc[np.arange(0+i, 500, 20)]

  min_n_receivers = 4 # min_number of receivers for a detection to be localized
  max_receiver_dist = 200  # maximum distance between receivers, set to all recievers
  cc_threshold = 0.001
  bandpass_ranges = {
    "riroriro": [2200, 5800],
    "kea": [2000, 6000]
    }
  cc_filter = "phat"

  tarray = SynchronizedRecorderArray(aru_coords, speed_of_sound=postemp.iloc[i]["SoS"])

  position_estimate = tarray.localize_detections(
      detection, min_n_receivers=min_n_receivers, max_receiver_dist=max_receiver_dist,
      cc_threshold=cc_threshold, bandpass_ranges=bandpass_ranges, cc_filter=cc_filter
  )

  results[i] = position_estimate[0]

  for j in position_estimate:
    if j.residual_rms < results[i].residual_rms:
      results[i] = j
```
